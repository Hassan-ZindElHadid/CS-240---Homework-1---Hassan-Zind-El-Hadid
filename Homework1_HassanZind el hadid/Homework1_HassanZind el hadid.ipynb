```python
from __future__ import print_function, division
import numpy as np
import pandas as pd
import thinkstats2
import thinkplot
import nsfg
```

```python
resp = nsfg.ReadFemResp()
resp.head(20)
# First 20 rows
```

```python
resp.tail(30)
# Last 30 rows
```

```python
rows = len(resp.index)
columns = len(resp.columns)
print("In this dataframe, there are "+str(rows)+" rows and "+str(columns)+" columns.")
```

```python
agescrn = resp.agescrn
agescrn
# agescrn column. Age of respondents
```

```python
resp.agescrn.min()
# Min age
```

```python
resp.agescrn.max()
# Max age
```

```python
resp[resp.agescrn <= 25]
# Respondents whose ages are less than or equal to 25 years old
```

```python
resp[resp.agescrn <= 25].numpregs.mean()
# Average of pregnancies in the lifetime of respondents of age equal to 25 years old or less
```

```python
fmarno = resp.fmarno
fmarno
# Column fmarno. Respondents' number of marriages
```

```python
resp.fmarno.value_counts().sort_index()
# Frequencies of number of marriages
```

```python
hist = thinkstats2.Hist(resp.fmarno, label = "Frequency of Number of Marriages")
thinkplot.Hist(hist, width = 0.5)
thinkplot.Config(xlabel = "Number of marriages", ylabel = "Frequency")
# Histogram showing number of marriages and their corresponding frequency for all respondents
```

```python
married = resp[resp.fmarno >= 1]
neverMarried = resp[resp.fmarno == 0]


married_hist = thinkstats2.Hist(married.totincr, label = "Married")
neverMarried_hist = thinkstats2.Hist(neverMarried.totincr, "Unmarried")

thinkplot.preplot(2)
thinkplot.Hist(married_hist, align = "right", width = 0.5)
thinkplot.Hist(neverMarried_hist, align = "left", width = 0.5)
thinkplot.Config(xlabel = "Income Category", ylabel = "Frequency of Married/Unmarried")

# Histogram showing frequency of Married/neverMarried with respect to income category (from column "totincr")
```

```python
# Min for married
min_married = married.totincr.min()
```

```python
# Max for married
max_married = married.totincr.max()
```

```python
# Standard deviation for married
std_married = married.totincr.std()
```

```python
# Mean for married
mean_married = married.totincr.mean()
```

```python
# Variance for married
var_married = married.totincr.var()
```

```python
# Min for neverMarried
min_neverMarried = neverMarried.totincr.min()
```

```python
# Max for neverMarried
max_neverMarried = neverMarried.totincr.max()
```

```python
# Standard deviation for neverMarried
std_neverMarried = neverMarried.totincr.std()
```

```python
# Mean for neverMarried
mean_neverMarried = neverMarried.totincr.mean()
```

```python
# Variance for neverMarried
var_neverMarried = neverMarried.totincr.var()
```

```python
# Stats for married respondents' total income
print("For married: max="+ str(max_married), " min="+str(min_married), " Standard deviation="+str(std_married), " Mean="+str(mean_married), " Variance="+str(var_married))
```

```python
# Stats for neverMarried respondents' total income
print("For neverMarried: max="+ str(max_neverMarried), " min="+str(min_neverMarried), " Standard deviation="+str(std_neverMarried), " Mean="+str(mean_neverMarried), " Variance="+str(var_neverMarried))
```
